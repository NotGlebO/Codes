---

#### Ja mums doti trƒ´s kontrolpunkti P0, P1, P2, tad mums nepiecie≈°ams:

> - 1) Noteikt sƒÅkumpunktu: tas ir pirmais kontrolpunkts P0.

> - 2) Noteikt soli t, ar kuru skaitƒ´s punktus

> - 3) Sakot ar t=0 izmantot t kƒÅ iteratƒ´vo soli, kamƒìr t = 1

> - 4) Noteikt Bern≈°teina polinomus solƒ´ t

> - 5) Noteikt punkta koordinƒÅtes solƒ´ t

> - 6) Attƒìlot uz ekrƒÅna punktu (x, y)

```python
import numpy as np               
import matplotlib.pyplot as plt  

img = np.ones((600,600,3))

#Man ir 3 kontrolpunkti 
                #P0    P1      P2     
def DrawBezier(x0,y0, x1,y1, x2,y2):
    
    x = x0 # sƒÅkumpunkta definƒì≈°ana pƒìc x ass
    y = y0 # sƒÅkumpunkta definƒì≈°ana pƒìc y ass
    t = 0 # parametrs t, kas pieder intervƒÅlam no 0 lƒ´dz 1
    ts = 0.001 # solis, ar kuru mainƒÅs parametrs t. Jo mazƒÅks ir solis ts,
    #jo vairƒÅk punktu uz lƒ´knes uzzƒ´mƒìs algoritms. 

    while t <= 1.0:  # kamƒìr nav sasniegts interavala gala punkts
        
        B0 = 1*(1-t) * (1-t) #rƒìƒ∑inam Ber≈°teina 0.polinomu
        B1 = 2* (1-t) * t #rƒìƒ∑inam Ber≈°teina 1.polinomu
        B2 = 1* t * t # #rƒìƒ∑inam Ber≈°teina 2.polinomu
       
        x = int(x0*B0 + x1*B1 + x2*B2) #rƒìƒ∑inam lƒ´knes pikseƒºa koordinƒÅtes pƒìc x ass
        y = int(y0*B0 + y1*B1 + y2*B2) #rƒìƒ∑inam lƒ´knes pikseƒºa koordinƒÅtes pƒìc y ass
        
        img[y,x] = (1,0,0) #zƒ´mƒìjam teko≈°o pikseli sarkanajƒÅ krƒÅsƒÅ
        
        t=t+ts #palielinam t vƒìrtƒ´bu
        
    return
             # P0         P1          P2          
DrawBezier(500, 100,     10, 10,    500, 550) #zƒ´mƒìjam lƒ´kni
    
%matplotlib notebook
plt.figure(figsize=(10,8), dpi = 100, facecolor='w')
plt.imshow(img)
plt.show()
```
