## DARBA UZDEVUMS
Individuālā uzdevuma izpildes laikā jārealizē pārvietošana, mērogošana un rotācija kādam 2D objektam, balstoties uz variantu, kuru var atrast ORTUS vidē. Variants atkarīgs no studentu apliecības numura pēdējiem 3 cipariem.

Uzdevuma variācijas: 
- pārvietot objektu par N-koordināšu skaitu, pa x un y asīm,
- mērogošana – samazināt vai palielināt objektu N-reizes pa x un y asīm, 
- rotēt objektu N-grādos.


```python
import math
import numpy as np
import matplotlib.pyplot as plt

img = np.ones((1000,1000,3))

point1 = (300,100)
point2 = (100,100)
point3 = (100,300)
point4 = (300,300)

    #Pārvietošana
def MoveObject(x1, y1, x2, y2, x3, y3, x4, y4):
   
    move_x = 400
    move_y = 200
    
    x1 = x1 + move_x
    y1 = y1 + move_y
    x2 = x2 + move_x
    y2 = y2 + move_y
    x3 = x3 + move_x
    y3 = y3 + move_y
    x4 = x4 + move_x
    y4 = y4 + move_y
    
    x5 = [x1,x2,x3,x4,x1]
    y5 = [y1,y2,y3,y4,y1]
    plt.plot(x5,y5)
    
    #Mērogošana
def ScaleObject(x1, y1, x2, y2, x3, y3, x4, y4):
    move_x = 2
    move_y = 4
    
    
    x1 = x1 * move_x
    y1 = y1 * move_y
    x2 = x2 * move_x
    y2 = y2 * move_y
    x3 = x3 * move_x
    y3 = y3 * move_y
    x4 = x4 * move_x
    y4 = y4 * move_y
    
    x5 = [x1,x2,x3,x4,x1]
    y5 = [y1,y2,y3,y4,y1]
    plt.plot(x5,y5)
    
    
    #Rotēšana
def RotateObject(x1, y1, x2, y2, x3, y3, x4, y4):
    #!Neaizmirst par prveidosanu no deg uz radians!!
    degree = 45
    a = math.radians(degree)
    xn1 = x1*math.cos(a) - y1*math.sin(a)
    yn1 = x1*math.sin(a) + y1*math.cos(a)
    xn2 = x2*math.cos(a) - y2*math.sin(a)
    yn2 = x2*math.sin(a) + y2*math.cos(a)
    xn3 = x3*math.cos(a) - y3*math.sin(a)
    yn3 = x3*math.sin(a) + y3*math.cos(a)
    xn4 = x4*math.cos(a) - y4*math.sin(a)
    yn4 = x4*math.sin(a) + y4*math.cos(a)
    
    x_points = [xn1,xn2,xn3,xn4,xn1]
    y_points = [yn1,yn2,yn3,yn4,yn1]
    plt.plot(x_points, y_points)
    

    
plt.figure(figsize=(10,8), dpi = 100, facecolor='w')


MoveObject(point1[0],point1[1],point2[0],point2[1],point3[0],point3[1],point4[0],point4[1])
ScaleObject(point1[0],point1[1],point2[0],point2[1],point3[0],point3[1],point4[0],point4[1])
RotateObject(point1[0],point1[1],point2[0],point2[1],point3[0],point3[1],point4[0],point4[1])


plt.show()
```
